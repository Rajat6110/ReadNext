# Problem Statement

Modern digital reading platforms contain large amounts of interaction data generated as users read books chapter by chapter. Unlike traditional recommendation systems where items are independent, reading behavior is highly sequential because chapters belong to ordered books and users typically progress through them in sequence.

The objective of this project is to build a hybrid recommendation system that can understand user reading behavior and generate meaningful personalized recommendations using historical interaction data and chapter metadata.

This project focuses on solving the following three recommendation problems:

### 1. Next Chapter Recommendation
Predict the next chapter a user is likely to read within books they are currently reading by utilizing sequential reading patterns and chapter ordering information.

### 2. Personalized Book Recommendation
Recommend new books to users based on their reading history, preferred genres, authors, and previously consumed content.

### 3. Cold Start Recommendation
Handle situations where:
- a new user has little or no interaction history
- a new book has very few interactions

by using popularity-based and content-based recommendation strategies.

# LAYER  -1 
### Problem 1 - Next Chapter Recommendation

The goal is to recommend the next chapter within all the books a user is currently reading

Since chapters are sequential, the recommendation is based on the latest chapter a user has read in a particular book.

```python
# sort interactions by reading order
sequence_data = df.sort_values(by=['user_id', 'book_id', 'chapter_sequence_no'])
sequence_data.head()
```

```python
# latest chapter read by user in each book
latest_chapters = sequence_data.groupby(['user_id', 'book_id'])['chapter_sequence_no'].max().reset_index()
latest_chapters.head()
```

```python
# predict next chapter sequence

latest_chapters['recommended_next_chapter'] = (
    latest_chapters['chapter_sequence_no'] + 1)
```

```python
# merge to get actual next chapter id

next_recommendations = latest_chapters.merge(
    df1[
        ['book_id', 'chapter_id', 'chapter_sequence_no']
    ],
    left_on=['book_id', 'recommended_next_chapter'],
    right_on=['book_id', 'chapter_sequence_no'],
    how='left'
)

# rename columns

next_recommendations = next_recommendations.rename(
    columns={
        'chapter_sequence_no_x': 'current_chapter',
        'chapter_id': 'recommended_chapter_id'
    }
)
```

```python
# remove completed books

next_recommendations = next_recommendations.dropna(
    subset=['recommended_chapter_id']
)
next_recommendations = next_recommendations.drop(
    columns=['chapter_sequence_no_y']
)
next_recommendations.head()
```

```python
# sample recommendations for one user

sample_user = next_recommendations['user_id'].iloc[0]
next_recommendations[
    next_recommendations['user_id'] == sample_user
][
    [
        'book_id',
        'current_chapter',
        'recommended_next_chapter'
    ]
]
```

## Approach

The recommendation system predicts the next unread chapter based on a user's reading progression within all books.

First, user interactions were sorted using:
- user_id
- book_id
- chapter_sequence_no

This helped reconstruct the reading sequence for each user.

Then, for every user-book pair, the latest chapter read by the user was identified.

The next chapter recommendation was generated using:

next chapter = current chapter + 1

Finally, the predicted chapter sequence was matched with the chapters dataset to retrieve the actual next chapter.

Books with no available next chapter were treated as completed and excluded from recommendations.

## Problem 2 - New Book Recommendation

Recommend new books to users based on their reading history and content preferences such as genres and authors.

```python
# for the 2nd problem we will be using the Content based Recommendation System , so that it can Recommend new book to user based upon past likings and interactions
```

```python
# content based rec
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python

# create metadata for each book

# create metadata for each book

books_data = df1.groupby(
    ['book_id', 'author_id']
)['tags'].apply(
    lambda x: ' '.join(set(x.astype(str)))
).reset_index()

books_data['metadata'] = (
    books_data['author_id'].astype(str) + ' ' +
    books_data['tags']
)

books_data.head()
```

```python
# convert metadata into tf-idf vectors

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(
    books_data['metadata']
)
```

```python
# mapping book ids to dataframe index

book_index = pd.Series(
    books_data.index,
    index=books_data['book_id']
).drop_duplicates()
```

```python
# recommend similar books

def recommend_books(book_id, top_n=10):

    idx = book_index[book_id]

    # compute similarity only for one book
    similarity_scores = cosine_similarity(
        tfidf_matrix[idx],
        tfidf_matrix
    ).flatten()

    similar_books = similarity_scores.argsort()[::-1][1:top_n+1]

    recommendations = books_data.iloc[
        similar_books
    ]['book_id']

    return books_data.iloc[
    similar_books
][
    ['book_id', 'tags']
]
```

```python
# sample recommendation - giving 10 rec to the user

sample_book = books_data['book_id'].iloc[15]

recommend_books(sample_book)
```

## Approach

This recommendation system suggests new books based on content similarity.

For each book, metadata was created by combining:
- author_id
- tags

The metadata was converted into numerical vectors using TF-IDF vectorization.

Cosine similarity was then used to measure similarity between books based on their genres, themes, and author patterns.

When a book is given as input, the system identifies books with the highest similarity scores and recommends them as potential next reads.

# Problem 3 - Cold Start Recommendation

Handle recommendation scenarios where users or books have little to no interaction history.

```python
# DURING EDA WE HAVE FIND - TOP POPULAR BOOKS , WE WILL USE THEM NOW FOR THE NEW USERS
```

```python
# recommend popular books for new users
# calculate book popularity


book_interactions = df2.groupby(
    'book_id'
).size().reset_index(name='interaction_count')

book_users = df2.groupby(
    'book_id'
)['user_id'].nunique().reset_index(
    name='unique_users'
)

popular_books = book_interactions.merge(
    book_users,
    on='book_id'
)

# engagement score

popular_books['engagement_score'] = (
    popular_books['interaction_count'] +
    popular_books['unique_users']
)

popular_books = popular_books.sort_values(
    by='engagement_score',
    ascending=False
)

popular_books.head()
```

```python
# most popular genres

popular_genres = df1['tags'].str.split(
    '|'
).explode().value_counts().reset_index()

popular_genres.columns = [
    'genre',
    'count'
]

popular_genres.head()
```

```python
# recommend popular books for new users

def recommend_popular_books(top_n=5):

    return popular_books.head(top_n)
```

```python
# sample cold start recommendation

recommend_popular_books() # this will be recommnended
```

To handle this cold start problem, a popularity-based recommendation strategy was used.

Books were ranked using:
- interaction counts
- overall engagement frequency

The most frequently engaged books are recommended as initial suggestions for new users.
